```python
df1 = pd.read_csv('/content/drive/MyDrive/ML_HANDS_ON/Data.csv')
df1.head()
```

Get Date by highest average sales

```python
hightest_sale = df1.groupby('Date')['Final Total'].mean().sort_values(ascending=False).head(1)
hightest_sale
```

```python
#How many total bills are there
total_bills = df1.groupby('Invoice No.')
len(total_bills)
```

```python
total_bills1 = df1['Invoice No.'].nunique()
print(total_bills1)
```

```python
#most sold item of September
september = df1[(df1['Date'] >= '2023-09-01') & (df1['Date'] <= '2023-09-30')]
september.groupby('Item Name')['Qty.'].sum().sort_values(ascending=False).head(1)
```

```python
#Sold Less in the Month of July
july = df1[(df1['Date'] >= '2023-07-01') & (df1['Date'] <= '2023-07-31')]
req = july.groupby('Category')['Qty.'].sum().sort_values(ascending=True).head(3)
```

```python
import seaborn as sns
sns.barplot(req)
```

```python
#Pie chart for slaes by category
sales = df1.groupby('Category')['Final Total'].sum()
sales
```

```python
import matplotlib.pyplot as plt
plt.pie(sales)
```

```python
#what is average order value of the restaurant
#AOV = total_amount / no.of.bills
total_amount = df1['Final Total'].sum()
no_of_bills = df1['Invoice No.'].nunique()
AOV = total_amount / no_of_bills
print(AOV)
```
