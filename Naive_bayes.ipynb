```python
#Import Libraries and DataSet
import pandas as pd
import numpy as np
weather = pd.read_csv('/content/weatherAUS.csv')
```

```python
weather.isna().sum()
```

```python
weather.shape
```

```python
# Drop columns with more than 10000 missing values
cols_to_drop = []
for i in weather.columns:
  if weather[i].isna().sum() > 10000:
    cols_to_drop.append(i)
weather = weather.drop(cols_to_drop, axis=1)
```

```python
# Fill missing values
for i in weather.columns:
  if weather[i].isna().sum() > 0:
    if weather[i].dtype == 'object' :
      weather[i] = weather[i].fillna(weather[i].mode()[0])
    else:
      weather[i] = weather[i].fillna(weather[i].mean())
```

```python
weather.isna().sum()
```

```python
weather.shape
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for i in weather:
  if weather[i].dtype == 'object':
    weather[i] = le.fit_transform(weather[i])
```

```python
x = weather.drop(['RainTomorrow', 'Date'], axis=1)
y = weather['RainTomorrow']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=32)
```

```python
from sklearn.naive_bayes import BernoulliNB
model = BernoulliNB()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```
